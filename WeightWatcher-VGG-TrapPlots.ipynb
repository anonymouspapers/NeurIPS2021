```python
plt.rcParams.update({'font.size': 20})
from pylab import rcParams
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (10,6.5),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'xx-large',
         'xtick.labelsize':'xx-large',
         'ytick.labelsize':'xx-large'}
plt.rcParams.update(params)
```

```python
def plot_loghist(x, bins, xmin):
    hist, bins = np.histogram(x, bins=bins)
    logbins = np.logspace(np.log10(bins[0]),np.log10(bins[-1]),len(bins))
    plt.hist(x, bins=logbins, density=True)

    if xmin:
        plt.axvline(xmin, color='r', label="xmin")

    plt.xscale('log')
```

### We set xmax, but we bget a little finger in the tail

```python
evals = esd_fc3[esd_fc3>0.01]
fc_layer = 'FC3'
bad_xmin = 5
```

### If we don't set xmax, we get crazy behavior

```python

fit = powerlaw.Fit(evals, xmax=np.max(evals))
fit2 = powerlaw.Fit(evals, xmax=np.max(evals), xmin=bad_xmin)


plot_loghist(evals[1:], bins=100, xmin=fit.xmin)
fig2 = fit.plot_pdf(color='b', linewidth=0)

fit.power_law.plot_pdf(color='red', linestyle='--', ax=fig2)
fit2.power_law.plot_pdf(color='darkviolet', linestyle='--', ax=fig2)

plt.axvline(x=fit.xmax, color='darkorange', label='xmax')
plt.axvline(bad_xmin, color='darkviolet', label='bad xmin')

plt.title("Log-Log ESD for VGG16 {}".format(fc_layer))
plt.legend()
plt.savefig("./paper/img/fig1-1a.png")
```

```python
plt.hist(evals, bins=100, density=True);
plt.title("Lin-Lin ESD for VGG16 {}".format(fc_layer))
plt.xlabel(r"$\lambda$")
plt.ylabel(r"$\rho(\lambda)$")
plt.axvline(fit.xmin, color='red', label='xmin')
plt.axvline(bad_xmin, color='darkviolet', label='bad xmin')
plt.axvline(x=fit.xmax, color='darkorange', label='xmax')

plt.legend()
plt.savefig("./paper/img/fig1-1b.png")
```

```python
 # plot log eigenvalue histogram                                                                                                                                           
nonzero_evals = evals[evals>0.0]
plt.title("Log-Lin ESD for VGG16 {}".format(fc_layer))
plt.hist(np.log10(nonzero_evals), bins=100, density=True)
#plt.title(r"Log10 ESD (Empirical Spectral Density) $\rho(\lambda)$" + "\nfor {} ".format(title))                                                                         
plt.axvline(x=np.log10(fit.xmin), color='red', label='xmin')
plt.axvline(x=np.log10(bad_xmin), color='darkviolet', label='bad min')
plt.axvline(x=np.log10(fit.xmax), color='darkorange', label='xmax')

plt.legend()
plt.savefig("./paper/img/fig1-1c.png")

```

```python

# plot xmins vs D                                                                                                                                                         
plt.plot(fit.xmins, fit.Ds, label=r'$D_{KS}$')
plt.axvline(x=fit.xmin, color='red', label='xmin')
plt.axvline(x=bad_xmin, color='darkviolet', label='bad xmin')

plt.title(r"KS Distance $(D_{KS})$ vs $\lambda_{min}=x_{min}$")

#plt.plot(fit.xmins, fit.sigmas / fit.alphas, label=r'$\sigma /\alpha$', linestyle='--')
plt.xlabel(r'$x_{min}$')
plt.ylabel(r'$D_{KS},\sigma,\alpha$')
#plt.title("current xmin={:0.3}".format(fit.xmin))                                                                                                                        
plt.legend()
plt.savefig("./paper/img/fig1-1d.png")
plt.show()
```
